### 11.4: Other Function Plots

#### Parametric Plots

```julia
plot(t->cos(t),t->sin(t),0,2*pi,legend=false)
```

```julia
plot(t->cos(t),t->sin(t),0,2*pi,aspect_ratio=:equal, legend=false)
```

#### Implicit Curve

```julia
contour(-1.1:0.01:1.1, -1.1:0.01:1.1, (x,y) -> x^2+y^2, levels=[1], aspect_ratio = :equal, colorbar_entry = false, tex_output_standalone = true)
```

```julia
savefig("test.tex")
```

```julia
pgfsave("circle.tex",pl,include_preamble=false)
```

```julia
contour(-10.1:0.1:10.1, -10.1:0.1:10.1, (x,y) -> sin(x+y)-cos(x*y)+1, levels=[0], aspect_ratio = :equal, colorbar_entry = false)
```

#### Surface Plots

```julia
f(x,y)=exp(-0.1*(x^2+y^2))
x = y = range(-5, stop = 5, length = 40)
surface(x,y,f)
```

#### Heat Maps

```julia
heatmap(x,y,f)
```
