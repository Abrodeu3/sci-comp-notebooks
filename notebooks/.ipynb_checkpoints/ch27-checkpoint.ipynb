```julia
function gradientDescent(f::Function,x0::Vector; γ = 0.25, max_steps = 100)
  local steps = 0
  local ∇0 = [1,1] # initialize it to get into while loop
  while norm(∇0)> 1e-8 && steps < max_steps
    ∇0 = ForwardDiff.gradient(f,x0)
    x0 = x0 - γ*∇0
    steps += 1
  end
  steps < max_steps || throw(ErrorException("The number of steps has exceeded $max_steps"))
  x0
end
```

```julia
gradientDescent(f,[0.1,0.1])
```

```julia
f([0,0])
```

```julia
function gradientDescentBB(f::Function,x₀::Vector)
  local steps = 0
  local ∇f₀ = ForwardDiff.gradient(f,x₀)
  local x₁ = x₀ - ∇f₀/norm(∇f₀)
  while norm(∇f₀)> 1e-4 && steps < max_steps
    ∇f₁ = ForwardDiff.gradient(f,x₁)
    Δ∇f = ∇f₁-∇f₀
    x₂ = x₁ - abs(dot(x₁-x₀,Δ∇f))/norm(Δ∇f)^2*∇f₁
    x₀ = x₁
    x₁ = x₂
    ∇f₀ = ∇f₁
  end
  steps < max_steps || throw(ErrorException("The number of steps has exceeded $max_steps"))
  x₁
end
```

```julia
gradientDescentBB(f,[0.1,0.1])
```
