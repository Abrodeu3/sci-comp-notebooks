## Chapter 9: Algorithm Analysis

```julia
function poly1(coeffs::Vector{T}, x::S) where {T <: Number, S <: Number}
  local sum = zero(T)
  function pow(x::T,n::Int) where T <: Number
    local prod = one(T)
    for j=1:n
      prod *=x
    end
  end 
  for n=1:length(coeffs)
    sum += coeffs[n]*pow(x,n-1)
  end
  sum
end
```

```julia
poly1([1,2,3,4],4)
```
